```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merged_data = pd.merge(mouse_metadata, study_results, on="Mouse ID")


# Display the data table for preview
merged_data.head()
```

```python
# Checking the number of mice.
unique_mice = merged_data['Mouse ID'].nunique()
print(f"Number of unique mice: {unique_mice}")
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_data = merged_data[merged_data.duplicated(['Mouse ID', 'Timepoint'])]
duplicate_mouse_id = duplicate_data['Mouse ID'].unique()[0]
print(f"Mouse ID with duplicate time points: {duplicate_mouse_id}")
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
print(duplicate_data)
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_data = merged_data[merged_data['Mouse ID'] != duplicate_mouse_id]
cleaned_data.head()
```

```python
# Checking the number of mice in the clean DataFrame.

# Drop the specific mouse ID from the DataFrame

# Display the updated number of unique mice IDs
unique_mice_cleaned = cleaned_data['Mouse ID'].nunique()
print(f"Number of unique mice after dropping specific mouse ID: {unique_mice_cleaned}")
```

## Summary Statistics

```python
# Calculate the mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
mean_volume = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median_volume = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
var_volume = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
std_volume = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem_volume = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

# Create a DataFrame to store the summary statistics
summary_stats = pd.DataFrame({
    'Mean Tumor Volume': mean_volume,
    'Median Tumor Volume': median_volume,
    'Tumor Volume Variance': var_volume,
    'Tumor Volume Std. Dev.': std_volume,
    'Tumor Volume Std. Err.': sem_volume
})

# Format the summary statistics DataFrame
summary_stats_formatted = summary_stats.style.format({
    'Mean Tumor Volume': '{:.6f}',
    'Median Tumor Volume': '{:.6f}',
    'Tumor Volume Variance': '{:.6f}',
    'Tumor Volume Std. Dev.': '{:.6f}',
    'Tumor Volume Std. Err.': '{:.6f}'
})

# Display the formatted summary statistics
display(summary_stats_formatted)
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
summary_stats = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])
summary_stats = summary_stats.rename(columns={
    'mean': 'Mean Tumor Volume',
    'median': 'Median Tumor Volume',
    'var': 'Tumor Volume Variance',
    'std': 'Tumor Volume Std. Dev.',
    'sem': 'Tumor Volume Std. Err.'
})

summary_stats_formatted = summary_stats.style.format({
    'Mean Tumor Volume': '{:.6f}',
    'Median Tumor Volume': '{:.6f}',
    'Tumor Volume Variance': '{:.6f}',
    'Tumor Volume Std. Dev.': '{:.6f}',
    'Tumor Volume Std. Err.': '{:.6f}'
})


# If you want to print it, you can use the following instead:
summary_stats_formatted
```

## Quartiles, Outliers and Boxplots

```python
# Calculate final tumor volume for each mouse across four treatment regimens
treatment_regimens = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
final_tumor_volumes = []

for regimen in treatment_regimens:
    final_volume = cleaned_data.loc[cleaned_data['Drug Regimen'] == regimen].groupby('Mouse ID')['Tumor Volume (mm3)'].last()
    final_tumor_volumes.append(final_volume)

# Determine outliers
outliers = {}
for i, regimen in enumerate(treatment_regimens):
    quartiles = final_tumor_volumes[i].quantile([0.25, 0.5, 0.75])
    lower_bound = quartiles[0.25] - 1.5 * (quartiles[0.75] - quartiles[0.25])
    upper_bound = quartiles[0.75] + 1.5 * (quartiles[0.75] - quartiles[0.25])
    outliers[regimen] = final_tumor_volumes[i][(final_tumor_volumes[i] < lower_bound) | (final_tumor_volumes[i] > upper_bound)]
```

```python


# Print outliers
print("Outliers:")
for regimen, outlier in outliers.items():
    print(f"{regimen}: {list(outlier.index)}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
# Create box plot
plt.figure(figsize=(8, 6))
plt.boxplot(final_tumor_volumes, labels=treatment_regimens, notch=True, patch_artist=True, flierprops={'marker': 'o', 'markerfacecolor': 'red', 'markersize': 8})
plt.xlabel('Drug Regimen')
plt.ylabel('Final Tumor Volume (mm3)')
plt.title('Distribution of Final Tumor Volume for Treatment Regimens')
plt.show()
```
